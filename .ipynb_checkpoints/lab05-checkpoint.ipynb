# Part 1

```python
def integrate(y, dx):
    """
    Takes an array of y-values and a step size dx of the corresponding x values 
    and numerically integrates the function, returning the final value
    """
    counter = 0
    for yi in y:
        counter += dx * yi
    return counter

```

```python
def plot_fn(xarr, yarr, title=''):
    """
    Basic plotting function
    """
    plt.plot(xarr, yarr)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
# Make Array
newAr = np.arange(0, 2*np.pi, 0.01)

# Begin plotting
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.plot(newAr, np.sin(newAr))
```

```python
# Calculate integrated value
intProd = integrate(np.sin(newAr), 0.01)
print (intProd)

newAr1 = np.arange(0, 2*np.pi, 0.001)
intProd1 = integrate(np.sin(newAr1), 0.001)
print (intProd1)

newAr2 = np.arange(0, 2*np.pi, 0.0000001)
intProd2 = integrate(np.sin(newAr1), 0.0000001)
print (intProd2)
```

```python
result = np.trapz(np.sin(newAr), dx = 0.01)
print (result)
```

# Part 2

```python
def wavepacket(x, k, sigma):
    """
    This function creates a wavepacket on the interval defined by x with
    wave Vector k and standard deviation sigma.
    """
    return np.sin(k*x) *  np.exp(-(x**2)/(2*sigma**2))
```

```python
def noisy_packet(x_values, k, sigma, noise_amplitude):
    """
    This function returns a noisy Gaussian wavepacket with wave
    vector k, standard deviation sigma and Gaussian noise of standard
    deviation noise_amplitude.
    """
    clean_y = wavepacket(x_values,k,sigma)
    noisy_y = clean_y + noise_amplitude*np.random.randn(len(x_values))
    return noisy_y
```

```python
# Generate and plot noisy waves
wave1 = noisy_packet(np.arange(0, 2*np.pi, 0.01) , 10, 3, 0.5)
wave2 = noisy_packet(np.arange(0, 2*np.pi, 0.01) , 10, 3, 0.1)
wave3 = noisy_packet(np.arange(0, 2*np.pi, 0.01) , 10, 100, 0.02)
wave4 = noisy_packet(np.arange(0, 2*np.pi, 0.01) , 5, 100, 0.02)
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.plot(np.arange(0, 2*np.pi, 0.01), wave1, 'green')
myplot.plot(np.arange(0, 2*np.pi, 0.01), wave2, 'orange')
myplot.plot(np.arange(0, 2*np.pi, 0.01), wave3, 'blue')
myplot.plot(np.arange(0, 2*np.pi, 0.01), wave4, 'red')
```

```python
def clean_data(x_values,y_values):
    """
    This function should take a set of y_values, perform the Fourier
    transform on it, filter out the high frequency noise, transform the
    signal back into real space, and return it.
    """

    # TODO edit this function
    y_fft = np.fft.rfft(y_values)
    
    low_pass_filter = np.ones(y_fft.shape) # build low pass filter for fourier function
    num_freq = y_fft.size
    low_pass_filter[int(num_freq/25):num_freq] = 0
    
    clean_data_fft = np.zeros(y_fft.shape)
    for idx, val in enumerate(low_pass_filter):
        if val == 1:
            clean_data_fft[idx] = y_fft[idx]
    
    y_clean_values = np.fft.irfft(clean_data_fft, len(y_values))
    
    return y_clean_values
```

```python
# TODO call noisy_packet() to get a Gaussian wave packet, 
# call clean_data() to apply a low pass filter to the data, and
# finally plot the result using plot_fn
wave = noisy_packet(np.arange(0, 2*np.pi, 0.01) , 10, 3, 0.1)
cleanWave = clean_data(np.arange(0, 2*np.pi, 0.01), wave)
fig = plt.figure()
myplot = fig.add_subplot(111)
myplot.plot(np.arange(0, 2*np.pi, 0.01), wave, 'green')
myplot.plot(np.arange(0, 2*np.pi, 0.01), cleanWave, 'orange')
```
